```python
#BY MS DHONI STILL ENOUGH TO BAT BY SHIKHAR BHATNAGAR
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("IPL_ball_by_ball_updated.csv")
```

```python
df.head()
```

```python
df['batting_team'].unique()
```

```python
df[df.batting_team == 'Chennai Super Kings']['striker'].unique()
```

```python
df_newseasons = df[df.season >= 2020]
df_newseasons.head(50)
```

```python
dhoni_df = df_newseasons[df_newseasons.striker == 'MS Dhoni']
```

```python
dhoni_df.head(2)
```

```python
print('Runs Scored :', dhoni_df['runs_off_bat'].sum())  # Runs scored
print('Balls Faced :', len(dhoni_df['ball']))  # Balls Faced
print('Strike rate :', 100*dhoni_df['runs_off_bat'].sum()/len(dhoni_df['ball']))  # Strike Rate
print('Average :', dhoni_df['runs_off_bat'].sum()/dhoni_df['player_dismissed'].count())  #Average
```

```python
for i in range(2020,2023,1):
    print('Strike rate in',i, ':', 100*dhoni_df[dhoni_df.season == i]['runs_off_bat'].sum()/len(dhoni_df[dhoni_df.season == i]['ball']))
```

```python
df_alldata1 = pd.DataFrame(df_newseasons.groupby(['striker'])['runs_off_bat'].sum().sort_values(ascending = False)).reset_index()
```

```python
df_alldata2 = pd.DataFrame(df_newseasons.groupby(['striker'])['ball'].count()).reset_index()
```

```python
df_runs = df_alldata1.merge(df_alldata2, on = 'striker', how = 'left')
df_runs
```

```python
df_runs['Strike_Rate'] = 100*df_runs['runs_off_bat']/df_runs['ball']
df_runs
```

```python
df_runs = df_runs[df_runs['ball'] >= 200]
df_runs
```

```python
plt.figure(figsize = (16,8))
plt.scatter(df_runs.runs_off_bat,df_runs.Strike_Rate)
for i in range(len(df_runs)):
    if df_runs['striker'][i] == 'MS Dhoni':
        plt.text(df_runs['runs_off_bat'][i], df_runs['Strike_Rate'][i], df_runs['striker'][i],color = 'red')
    else:
        plt.text(df_runs['runs_off_bat'][i] , df_runs['Strike_Rate'][i] , df_runs['striker'][i])
plt.title("Batsman Runs Scored between 2020 - 2022", fontsize = 25)
plt.xlabel("Runs Scored", fontsize = 16)
plt.ylabel("Strike Rate", fontsize = 16)
plt.axhline(130,ls = '--', color = 'grey')
plt.axvline(900,ls = '--', color = 'grey')
plt.show()
```
